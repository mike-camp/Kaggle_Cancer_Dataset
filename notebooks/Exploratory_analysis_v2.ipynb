```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
def return_same(x):
    return x
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import log_loss
pipe = Pipeline([('encoder',HashingVectorizer(analyzer=return_same)),
                 ('lr',xgboost.XGBClassifier(n_jobs=-1))])
param_grid = {'lr__max_depth':[2,3,4]}
gs = GridSearchCV(pipe, param_grid, scoring = 'neg_log_loss',n_jobs=1,cv=5)
gs.fit(X.processed,X.Class)
print(gs.best_score_)
gs.grid_scores_
```

```python
pipe = Pipeline([('encoder',HashingVectorizer(analyzer=return_same)),
                 ('lr',xgboost.XGBClassifier(n_jobs=-1,learning_rate=.01,n_estimators=300))])
param_grid = {'lr__max_depth':[2]}
gs = GridSearchCV(pipe, param_grid, scoring = 'neg_log_loss',n_jobs=1,cv=5)
gs.fit(X.processed,X.Class)
print(gs.best_score_)
gs.grid_scores_
```
